```python
df=df.drop(['Neo Reference ID','Name','Equinox'],axis=1)
```

```python
df=df.drop(['Orbiting Body'],axis=1)
df=df.drop(['Close Approach Date'],axis=1)
df=df.drop(['Orbit Determination Date'],axis=1)
```

```python
df['Hazardous']=df['Hazardous'].replace(['False', 'True'], [1,0])
```

```python
x=df.drop(['Hazardous'],axis=1)
y=df['Hazardous'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,confusion_matrix,classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g=GaussianNB()
b=BernoulliNB()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=.20, random_state=42)
```

```python
g.fit(x,y)  
```

```python
b.fit(x,y)
```

```python
tahmin1=g.predict(x)
```

```python
accuracy_score(tahmin1,y)
```

```python
tahmin2=b.predict(x)
```

```python
accuracy_score(tahmin2,y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
h=GradientBoostingClassifier()
h.fit(x,y)
tahmin4=h.predict(x)
accuracy_score(tahmin4,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
tahmin3 = KNeighborsClassifier()
tahmin3.fit(x_train,y_train)
tahmin3TrainScore = round(tahmin3.score(x_train,y_train)*100,2)
tahmin3TestScore = round(tahmin3.score(x_test,y_test)*100,2)
print("Model's training accuracy = " + str(tahmin3TrainScore)+"%")
print("Model's test accuracy = " + str(tahmin3TestScore)+"%")
```
